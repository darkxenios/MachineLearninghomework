Этот набор данных предоставляет информацию о клиентах для прогнозирования выбытия клиентов по потребительским кредитным картам.

CLIENTNUM: Уникальный идентификатор для каждого клиента. (Целое число)
Attrition_Flag: Флаг, указывающий, закончился ли клиент. (Логическое значение)
Customer_Age: Возраст клиента. (Целое число)
Gender: Пол клиента. (Строка)
Dependent_count: Количество иждивенцев, которые есть у клиента. (Целое число)
Education_Level: уровень образования клиента. (Строка)
Marital_Status: Семейное положение клиента. (Строка)
Income_Category: Категория дохода клиента. (Строка)
Card_Category: Тип карты, которой владеет клиент. (Строка)
Months_on_book: Как долго клиент числится в бухгалтерских книгах. (Целое число)
Total_Relationship_Count: Общее количество отношений клиента с поставщиком кредитной карты. (Целое число)
Months_Inactive_12_mon: Количество месяцев, в течение которых клиент был неактивен за последние двенадцать месяцев. (Целое число)
Contacts_Count_12_mon: Количество контактов, которые были у клиента за последние двенадцать месяцев. (Целое число)
Credit_Limit: Кредитный лимит клиента. (Целое число)
Total_Revolving_Bal: Общий возобновляемый баланс клиента. (Целое число)
Avg_Open_To_Buy: Среднее соотношение открытых и закрытых покупок клиента. (Целое число)
Total_Amt_Chng_Q4_Q1: Общая сумма изменилась с четвертого по первый квартал. (Целое число)
Total_Trans_Amt: Общая сумма транзакции. (Целое число)
Total_Trans_Ct: Общее количество транзакций. (Целое число)
Total_Ct_Chng_Q4_Q1: Общее количество изменилось с четвертого квартала на первый. (Целое число)
Avg_Utilization_Ratio: Средний коэффициент использования клиентов. (Целое число)
